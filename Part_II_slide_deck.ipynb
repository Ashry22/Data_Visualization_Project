## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from joypy import joyplot

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
trip_data= pd.read_csv('201902-fordgobike-tripdata-modified.csv')
```

## 1. What is the average usage duration?

> Increasing the average duration of usage is an indicator that customer satisfy with our products and depend on it.


```python
plt.figure(figsize=(12, 5))

sb.histplot(trip_data.query('trip_type == "short_trip"').duration_min, bins=40)
plt.xlabel('Durations (min)')
plt.ylabel('No. of customers')
plt.title('Relation between Durations and No. of Customers');
```

## 2. Is there areas should be improve to increase this average duration?

## Yes, and the first area to be investaged is the customer retention over the days of the week.

> The customer retention decreases during holidays. So, marketing campaign and promotions for holidays would be a great idea.

```python
plt.figure(figsize=(12, 5))

week_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
color_base = sb.color_palette()[0]
sb.countplot(data=trip_data, x='start_weekday', order=week_days, color=color_base)
plt.xticks(rotation=15)

plt.xlabel('Day of the Week')
plt.ylabel('No. of customers')
plt.title('Relation between Day of the Week and No. of Customers at Short Trips');
```

## b. The second area of investigation is the a relation between subscribers and customers behavior over the days of the week.


```python
plt.figure(figsize=(12, 5))

sb.violinplot(data=trip_data.query('trip_type == "short_trip"'), x='start_weekday', y='duration_min', 
              hue='user_type', split=True)
plt.xticks(rotation=15);
plt.xlabel('Day of the Week')
plt.ylabel('Duration (min)')
plt.title('Relation between Day of the Week & Duration at Short Trips');
```

```python
plt.figure(figsize=(12, 5))

sb.violinplot(data=trip_data.query('trip_type == "long_trip"'), x='start_weekday', y='duration_hour', 
              hue='user_type', split=True)
plt.xticks(rotation=15);
plt.xlabel('Day of the Week')
plt.ylabel('Duration (min)')
plt.title('Relation between Day of the Week & Duration at Long Trips');
```

> As we can see, Customer types Customers tend to use bike over longer duration at Saturdays and Sundays at short trips.
So, promotion for long trips would make some improvements.
